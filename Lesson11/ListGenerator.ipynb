```python
import matplotlib.pyplot as plt

def plot_list(the_list):
    plt.plot(the_list, 'ro')
    plt.xlabel("numer pozycji")
    plt.ylabel("liczba na pozycji")
    plt.show()
```

```python
import random

def generate_random_list(number_of_elements):
    the_list = list(range(number_of_elements - 1))
    random.shuffle(the_list)
    return the_list

my_list = generate_random_list(100)
plot_list(my_list)
plot_list(sorted(my_list))
```

```python
import math

def generate_almost_sorted_list(number_of_elements, noise_ratio=1):
    resulting_list = []
    swapping_region = int(noise_ratio * math.log(number_of_elements, 2))
    if swapping_region >= number_of_elements:
        swapping_region = int(math.log(number_of_elements, 2))
    i = 0
    while len(resulting_list) < number_of_elements:
        if (i + swapping_region >= number_of_elements):
            sublist = list(range(i, number_of_elements))
        else:
            sublist = list(range(i, i + swapping_region))
        random.shuffle(sublist)
        resulting_list.extend(sublist)
        i += swapping_region
    return resulting_list

my_list = generate_almost_sorted_list(100, noise_ratio=2.5)
plot_list(my_list)
plot_list(sorted(my_list))
```

```python
def generate_almost_reversely_sorted_list(number_of_elements, noise_ratio=1):
    resulting_list = []
    swapping_region = int(noise_ratio * math.log(number_of_elements, 2))
    if swapping_region >= number_of_elements:
        swapping_region = int(math.log(number_of_elements, 2))
    i = number_of_elements - 1
    while i >= 0:
        if (i - swapping_region < 0):
            sublist = list(range(i, -1, -1))
        else:
            sublist = list(range(i, i - swapping_region, -1))
        random.shuffle(sublist)
        resulting_list.extend(sublist)
        i -= swapping_region
    return resulting_list

my_list = generate_almost_reversely_sorted_list(100)
plot_list(my_list)
plot_list(sorted(my_list))
```

```python
import numpy

def generate_gaussian_distribution_list(number_of_elements):
    return numpy.random.normal(size=number_of_elements)

my_list = generate_gaussian_distribution_list(100)
plot_list(my_list)
plot_list(sorted(my_list))
```

```python
def generate_n_numbers_from_k_element_set(number_of_elements, k_element_set):
    the_list = [None] * number_of_elements
    for i in range(number_of_elements):
        the_list[i] = k_element_set[random.randint(0, len(k_element_set) - 1)]
    return the_list

my_list = generate_n_numbers_from_k_element_set(100, [-11, -7, -5, -3, 3, 5, 7, 11])
plot_list(my_list)
plot_list(sorted(my_list))
```
